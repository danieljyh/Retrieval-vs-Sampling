```python
import json
import matplotlib.pyplot as plt

with open("../data/qaego4d/test.json", "r") as f:
    dataset = json.load(f)

frame_lens = []
for data in dataset:
    for sample in data["conversations"]:
        frame_st, frame_ed = sample["temporal_windows"][0]
        frame_lens.append(frame_ed - frame_st)

bins = range(0, 100, 5)
plt.hist(frame_lens, bins=bins)
plt.show()
```
